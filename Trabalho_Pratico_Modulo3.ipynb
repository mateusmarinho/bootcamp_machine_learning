##**Modelo de regressão linear: diabetes_numeric.csv**

```
# separando os conjuntos de treino e teste
X = np.array(df_diabetes[['age', 'deficit']])
y = np.array(df_diabetes['c_peptide'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```
# importando o moddelo de regressão linear
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
```

```
# treinando o modelo com o conjunto de treino
regression.fit(X_train, y_train)
```

```
# aplicando o modelo treinado para previsão
y_pred = regression.predict(X_test)
```

```
p = pd.DataFrame({'Real': y_test, 'Previsto': y_pred})
p.head(10)
```

```
# avaliando o modelo
print('R2: ', metrics.r2_score(y_test, y_pred))
print('MAE: ', metrics.mean_absolute_error(y_test, y_pred))
print('MSE: ', metrics.mean_squared_error(y_test, y_pred))
```

##**Modelo de classificação: bloodtransf.csv**

```
# separando os conjuntos de treino e teste
X = np.array(df_blood[['V1', 'V2', 'V3', 'V4']])
y = np.array(df_blood['Class'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```
# importando o modelo SVM
from sklearn.svm import SVC

classifier = SVC(kernel='rbf')
```

```
# treinando o modelo SVM
classifier.fit(X_train, y_train)
```

```
# aplicando o modelo treinado para previsão
y_pred = classifier.predict(X_test)
```

```
p = pd.DataFrame({'Real': y_test, 'Previsto': y_pred})
p.head(10)
```

```
# avaliando o modelo
print('Acurácia: ', metrics.accuracy_score(y_test, y_pred))
print('Precision: ', metrics.precision_score(y_test, y_pred))
print('Recall: ', metrics.recall_score(y_test, y_pred))
print('F1: ', metrics.f1_score(y_test, y_pred))
print('AUROC: ', metrics.roc_auc_score(y_test, y_pred))
print('Matriz de confusão:\n', metrics.confusion_matrix(y_test, y_pred))
```

```
# imprimindo relatório de métricas de qualidade
print('Classification Report:\n', metrics.classification_report(y_test, y_pred))
```

##**Modelo de clusterização: wine.csv**

```
# criando um dicionário de dados para o mapeamento
classes = {1: 0, 2: 1, 3: 2}

#substituindo os valores categóricos na coluna 'class' pelo mapeamento
df_wine['class'] = df_wine['class'].map(classes)

# verificando
df_wine.head(10)
```

```
# separando os conjuntos de treino e teste
y = np.array(df_wine['class'])
df_wine = df_wine.drop('class', axis = 1)
X = np.array(df_wine)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```
# importando o modelo K-means
from sklearn.cluster import KMeans

clustering = KMeans(n_clusters = 3, random_state = 42)
```

```
# treinando o modelo com o conjunto de treino
clustering.fit(X_train)
```

```
# aplicando o modelo treinado para previsão
y_pred = clustering.predict(X_test)
```

```
p = pd.DataFrame({'Real': y_test, 'Previsto': y_pred})
p.head(10)
```

```
# avaliando o modelo
print('Coeficiente de silhueta: ', metrics.silhouette_score(X_test, y_pred))
print('Davies-Bouldin Score: ', metrics.davies_bouldin_score(X_test, y_pred))
print('Mutual information: ', metrics.mutual_info_score(y_test, y_pred))
```
